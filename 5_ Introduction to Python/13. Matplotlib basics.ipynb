```python
import matplotlib.pyplot as plt
%matplotlib inline
#%matplotlib notebook
```

```python
import keras
from keras.datasets import mnist
```

```python
#from IPython.core.interactiveshell import InteractiveShell
#InteractiveShell.ast_node_interactivity="all"
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
x_train.shape
y_train.shape
y_test.shape
x_test.shape
```

```python
plt.figure(figsize=(3,3))
plt.imshow(x_train[0],'gray')
```

```python
plt.subplot(1,3,1).imshow(x_train[0],'gray')
plt.subplot(1,3,2).imshow(x_train[100],'gray')
plt.subplot(1,3,3).imshow(x_train[1000],'gray')
plt.show()
```

```python
for i in range(3):
    plt.subplot(1,3,i+1).imshow(x_train[i],'gray')
plt.show()
```

```python
import numpy as np
x = np.linspace(-np.pi,np.pi,7)

c = np.cos(x)
d = np.sin(x)

plt.plot(x,c)
plt.plot(x,d)
```

```python
import numpy as np
x = np.linspace(-np.pi,np.pi,256)

c = np.cos(x)
d = np.sin(x)

plt.plot(x,c)
plt.plot(x,d)
```

```python
import numpy as np
x = np.linspace(-np.pi,np.pi,256)

c = np.cos(x)
d = np.sin(x)

plt.plot(x,c,color="red",linewidth=3,linestyle="--")
plt.plot(x,d,color="black",linewidth=3,linestyle="--")
```
